```python
#Changing the column names for df2
df2.columns=['DR_NO', 'Date Rptd', 'DATE OCC', 'TIME OCC', 'AREA ', 'AREA NAME',
       'Rpt Dist No', 'Part 1-2', 'Crm Cd', 'Crm Cd Desc', 'Mocodes',
       'Vict Age', 'Vict Sex', 'Vict Descent', 'Premis Cd', 'Premis Desc',
       'Weapon Used Cd', 'Weapon Desc', 'Status', 'Status Desc', 'Crm Cd 1',
       'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'LOCATION', 'Cross Street', 'LAT',
       'LON']
```

```python
#Concatenating both the data sets
k=pd.concat([df,df2],ignore_index=True)
```

```python
#We need to convert date rptd and date occ into date time.
#We have made date occ as the index of this dataset
k["Date Rptd"]=pd.to_datetime(k["Date Rptd"])
k["DATE OCC"]=pd.to_datetime(k["DATE OCC"])
k.set_index(["DATE OCC"],inplace=True)
k["count"]=1
k
k.info()
```

```python
#To sum it yearwise
Yearwise_Count=k["count"].resample('Y').sum()
Yearwise_Count
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

Yearwise_Count.plot(kind='bar',title="crime occuring year-wise")
```
