```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Lakshi\Downloads\Crime_Data_from_2010_to_2019.csv")
df2=pd.read_csv(r"C:\Users\Lakshi\Downloads\Crime_Data_from_2020_to_Present.csv")
```

```python
df.columns
df2.columns=['DR_NO', 'Date Rptd', 'DATE OCC', 'TIME OCC', 'AREA ', 'AREA NAME',
       'Rpt Dist No', 'Part 1-2', 'Crm Cd', 'Crm Cd Desc', 'Mocodes',
       'Vict Age', 'Vict Sex', 'Vict Descent', 'Premis Cd', 'Premis Desc',
       'Weapon Used Cd', 'Weapon Desc', 'Status', 'Status Desc', 'Crm Cd 1',
       'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'LOCATION', 'Cross Street', 'LAT',
       'LON']
```

```python
k=pd.concat([df,df2],ignore_index=True)
k1=k.copy()
k1.columns
k2=k1[['DR_NO', 'DATE OCC', 'TIME OCC', 'AREA ','AREA NAME',
       'Rpt Dist No', 'Crm Cd', 'Crm Cd Desc', 'Mocodes',
       'Vict Age', 'Vict Sex', 'Premis Desc',
       'Weapon Used Cd','Weapon Desc', 'Status','Status Desc', 'LOCATION']]
```

```python
#Changing the data type
k2['DATE OCC']=k2['DATE OCC'].astype(str).str.replace('-','/')
k2['DATE OCC']=pd.to_datetime(k2['DATE OCC'])
k2['TIME OCC']=k2['TIME OCC']/100
```

```python
k2['DATE OCC']=k2['DATE OCC'].astype(str)+' '+k2['TIME OCC'].astype(str)
```

```python
k2['DATE OCC']
```

```python
k2['DATE OCC']=pd.to_datetime(k2['DATE OCC'],format='%Y-%m-%d %H.%M')
k2['DATE OCC']
#Now we have time along the date here
```

```python
k2.index=pd.DatetimeIndex(k2['DATE OCC'])
k2['HOUR']=k2['DATE OCC'].apply(lambda x:x.hour)
k2['MONTH']=k2['DATE OCC'].apply(lambda x:x.month)
k2['YEAR']=k2['DATE OCC'].apply(lambda x:x.year)
```

```python
#how number of crimes vary in each month across years
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))

k2.resample('M').size().plot(legend=False)
plt.title('Number of crimes per month (2010-2021)')
plt.xlabel('Months')
plt.xlabel('Number of crimes')
plt.show()
```

```python
import numpy as np
cr=k2.pivot_table('DR_NO',aggfunc=np.size,columns='Crm Cd Desc',index=k2.index,fill_value=0)
```

```python
m=k2['Crm Cd Desc'].value_counts()[:10]
cr10=list(m.index)
cr10
m
```

```python
k3=k2[k2['Crm Cd Desc'].isin(cr10)]
crbyhour=k3.groupby('HOUR').size().reset_index(name='No of crime')
crbyhour['HOUR']=crbyhour['HOUR'].apply(lambda x:str(x)+":00")
crbyhour
```

```python
#Number of crimes in each hour of the day
import seaborn as sns
fig=plt.figure(figsize=(15,10))

sns.pointplot(data=crbyhour, x="HOUR", y="No of crime", color="salmon")
plt.title('Number of crimes per month (2010-2021)')
plt.xlabel('Time')
plt.ylabel('Number of crimes')
plt.show()
```

```python
#Type of crimes(Top 10) as per each hour of the day
k4=k2[k2['Crm Cd Desc'].isin(cr10)]
crbytype=k4.groupby(['HOUR','Crm Cd Desc']).size().reset_index(name='No of crime')
crbytype['HOUR']=crbytype['HOUR'].apply(lambda x:str(x)+":00")
ccrbytype1=crbytype.pivot_table('No of crime',aggfunc=np.sum,columns=['Crm Cd Desc'],index=['HOUR'],fill_value=0)
ccrbytype1.plot(kind='bar',stacked=True,figsize=(16,12),title='No of crimes by types')
```

```python
k4=k2[k2['Crm Cd Desc'].isin(cr10)]
crbymonth=k3.groupby('MONTH').size().reset_index(name='No of crime')
crbymonth['avg']=crbymonth['No of crime'].apply(lambda x:x/11)
crbymonth
```

```python
fig=plt.figure(figsize=(15,10))

sns.pointplot(data=crbymonth, x="MONTH", y="avg", color="salmon")
plt.title('Number of crimes per month (2010-2021)')
plt.xlabel('Time')
plt.ylabel('Number of crimes')
plt.show()
```

```python
d={'Invest Cont':False,
 'Adult Other':False,
 'Adult Arrest':True,
 'Juv Arrest':True,
 'Juv Other':False,
 'UNK':False}
k2['Status Desc2']=k2['Status Desc'].map(lambda x: d[x])
```

```python
#graph showing if the accused was arrested or not in top 10 crimes
crbyarrest=k2.groupby(['Crm Cd Desc','Status Desc2']).size().reset_index(name='No of crime')
crbyarrest1=crbyarrest[crbyarrest['Crm Cd Desc'].isin(cr10)]
ccrbyarrest2=crbyarrest1.pivot_table('No of crime',aggfunc=np.sum,columns='Status Desc2',index='Crm Cd Desc',fill_value=0).reset_index()
ccrbyarrest2['Total']=ccrbyarrest2[False]+ccrbyarrest2[True]
ccrbyarrest2['y']=ccrbyarrest2[True]/ccrbyarrest2['Total']
ccrbyarrest2['N']=ccrbyarrest2[False]/ccrbyarrest2['Total']
ccrbyarrest2=ccrbyarrest2[['Crm Cd Desc','N','y']]
ccrbyarrest3=ccrbyarrest2.set_index('Crm Cd Desc')
ccrbyarrest3.plot(kind='barh',stacked=True,figsize=(16,6),title='No of crimes by types')
```
